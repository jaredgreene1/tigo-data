```python
# Setup
%matplotlib inline
import pandas as pd

DATA_FILE   = './data/Calls_20180426.csv' # Saturday
SAMPLE_SIZE = 1200000

data = pd.read_csv(DATA_FILE, nrows = SAMPLE_SIZE)
```

```python
# check this for drawing the graph https://stackoverflow.com/questions/19212979/draw-graph-in-networkx

import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()

G.add_edges_from(zip(data.ANUMBER.values, data.BNUMBER.values))
```

```python
data.head()
```

```python
G.number_of_nodes()/2
```

```python
cc = nx.connected_component_subgraphs(G)
```

```python
# Number of vertex disjoint subgraphs

cl = list(cc)
"{} disjointed graphs".format(len(cl))
```

```python
# % of calls in simple graphs

"{:0.1%} of calls are in simple graphs (graphs with a single edge)".format(len(cl)/SAMPLE_SIZE)
```

```python
nx.density(G)
```

```python
Gc = max(nx.connected_component_subgraphs(G), key=len)
Gc
```

```python
len(Gc.edges)
```
